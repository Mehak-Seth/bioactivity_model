```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
# Inspired by: https://codeocean.com/explore/capsules?query=tag:data-curation

def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem)
        moldata.append(mol)

    baseData= np.arange(1,1)
    i=0
    for mol in moldata:

        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)

        row = np.array([desc_MolWt,
                        desc_MolLogP, #solubility
                        desc_NumHDonors,
                        desc_NumHAcceptors])

        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1

    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)

    return descriptors
```

```python
# https://github.com/chaninlab/estrogen-receptor-alpha-qsar/blob/master/02_ER_alpha_RO5.ipynb

import numpy as np

def pIC50(input):
    pIC50 = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop('standard_value', axis=1)


    return x
```

```python
def mannwhitney(descriptor, verbose=False):
  # https://machinelearningmastery.com/nonparametric-statistical-significance-tests-in-python/
  from numpy.random import seed
  from numpy.random import randn
  from scipy.stats import mannwhitneyu

# seed the random number generator
  seed(1)

# actives and inactives
  selection = [descriptor, 'bioactivity_class']
  df = df_2class[selection]
  active = df[df.bioactivity_class == 'active']
  active = active[descriptor]

  selection = [descriptor, 'bioactivity_class']
  df = df_2class[selection]
  inactive = df[df.bioactivity_class == 'inactive']
  inactive = inactive[descriptor]

# compare samples
  stat, p = mannwhitneyu(active, inactive)
  #print('Statistics=%.3f, p=%.3f' % (stat, p))

# interpret
  alpha = 0.05
  if p > alpha:
    interpretation = 'Same distribution (fail to reject H0)'
  else:
    interpretation = 'Different distribution (reject H0)'

  results = pd.DataFrame({'Descriptor':descriptor,
                          'Statistics':stat,
                          'p':p,
                          'alpha':alpha,
                          'Interpretation':interpretation}, index=[0])
  filename = 'mannwhitneyu_' + descriptor + '.csv'
  results.to_csv(filename)

  return results
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Ridge
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Load the dataset
!wget https://github.com/dataprofessor/data/raw/master/acetylcholinesterase_06_bioactivity_data_3class_pIC50_pubchem_fp.csv
df = pd.read_csv('acetylcholinesterase_06_bioactivity_data_3class_pIC50_pubchem_fp.csv')

# Define features and target
X = df.drop('pIC50', axis=1)
Y = df['pIC50']

# Perform feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform data splitting using 80/20 ratio
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)

# Define models to evaluate
models = {
    "Gradient Boosting Regressor": GradientBoostingRegressor(),
    "XGBoost Regressor": xgb.XGBRegressor(),
    "Random Forest Regressor": RandomForestRegressor(),
    "Decision Tree Regressor": DecisionTreeRegressor(),
    "Support Vector Regressor": SVR(),
    "Ridge Regression": Ridge()
}

# Initialize performance records
train_results = {
    "Model": [],
    "R-Squared (Train)": [],
    "RMSE (Train)": [],
}

test_results = {
    "Model": [],
    "R-Squared (Test)": [],
    "RMSE (Test)": [],
}

# Evaluate each model
for model_name, model in models.items():
    try:
        # Fit the model
        model.fit(X_train, Y_train)

        # Make predictions
        Y_train_pred = model.predict(X_train)
        Y_test_pred = model.predict(X_test)

        # Calculate R-squared and RMSE for training set
        r_squared_train = r2_score(Y_train, Y_train_pred)
        rmse_train = np.sqrt(mean_squared_error(Y_train, Y_train_pred))

        # Calculate R-squared and RMSE for test set
        r_squared_test = r2_score(Y_test, Y_test_pred)
        rmse_test = np.sqrt(mean_squared_error(Y_test, Y_test_pred))

        # Record training results
        train_results["Model"].append(model_name)
        train_results["R-Squared (Train)"].append(r_squared_train)
        train_results["RMSE (Train)"].append(rmse_train)

        # Record testing results
        test_results["Model"].append(model_name)
        test_results["R-Squared (Test)"].append(r_squared_test)
        test_results["RMSE (Test)"].append(rmse_test)

    except Exception as e:
        print(f"Error with model {model_name}: {e}")

# Create DataFrames to store results
train_results_df = pd.DataFrame(train_results)
test_results_df = pd.DataFrame(test_results)

# Display training and testing results
print("Training Results:")
print(train_results_df)

print("\nTesting Results:")
print(test_results_df)

# Set the style for seaborn
sns.set(style="whitegrid")

# Plotting Training RMSE
plt.figure(figsize=(10, 5))
sns.barplot(y="Model", x="RMSE (Train)", data=train_results_df, palette="viridis")
plt.title("Training RMSE Values of Models", fontsize=16)
plt.xlabel("RMSE (Train)", fontsize=14)
plt.ylabel("Models", fontsize=14)
plt.xlim(0, train_results_df["RMSE (Train)"].max() + 1)  # Set x-limits
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plotting Testing RMSE
plt.figure(figsize=(10, 5))
sns.barplot(y="Model", x="RMSE (Test)", data=test_results_df, palette="plasma")
plt.title("Testing RMSE Values of Models", fontsize=16)
plt.xlabel("RMSE (Test)", fontsize=14)
plt.ylabel("Models", fontsize=14)
plt.xlim(0, test_results_df["RMSE (Test)"].max() + 1)  # Set x-limits
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
